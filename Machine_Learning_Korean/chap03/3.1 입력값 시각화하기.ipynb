```python
import numpy as np # 원데이터의 초기화를 위해 NumPy를 불러옵니다
import matplotlib.pyplot as plt # 데이터를 시각화하기 위해 matplotlib 라이브러리를 이용합니다
```

```python
x_train = np.linspace(-1, 1, 101) # 입력값은 -1과 1 사이에 균등하게 위치하는 101개의 값으로 구성합니다
y_train = 2 * x_train + np.random.randn(*x_train.shape) * 0.33 # 출력값은 대체로 입력값에 비례하지만 노이즈가 들어가게 구성합니다
```

```python
plt.scatter(x_train, y_train) # matplotlib 함수를 이용하여 데이터의 산포도를 생성합니다
plt.show()
```
